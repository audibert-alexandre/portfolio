## Imports

### Librairies

```python
import numpy as np
import pandas as pd
import random
import warnings

warnings.filterwarnings("ignore")
pd.set_option('display.max_rows', 100)
pd.set_option('display.max_columns', 60)

import matplotlib.pyplot as plt
import sys 

from datetime import datetime, date, timedelta

import lightgbm as lgb

from sklearn.metrics import r2_score
from scipy.stats import pearsonr
```

### Data

```python
train = pd.read_csv("D:/GitHub/Dataset/g-research-crypto-forecasting/train.csv")
```

### Data transformation

```python
# We change timestamp data to date (in seconde frequence)
first_step = 0
step = 1000000
df = pd.DataFrame()

for rows_interval in range(first_step, train.shape[0], step):
    tmp_df = train.iloc[rows_interval:rows_interval + step]
    tmp_df['datetime'] = tmp_df.timestamp.apply(datetime.fromtimestamp)
    df = pd.concat([df, tmp_df], axis=0)
```

### Train/Test & Validation Split

```python
# Creating a train_test & validation set 
split_date = '2021-06-18'
train_test_set = df.loc[df.datetime <= split_date].copy()
validation_set = df.loc[df.datetime > split_date].copy()
```

### Data Transformation

```python
#We replace eacch Target's null valu to its median
train_test_set['Target'] = train_test_set['Target'].fillna(train_test_set['Target'].median())
```

### Feature creation

```python
# Here we are splitting our dataset by Asset
l_train_test_set = []

for asset in range(0,14):
    l_train_test_set.append(train_test_set[train_test_set.Asset_ID == asset])
```

```python
# Here we are going to create, for each asset, 15 shifted columns from our Target 
for i in  range(len(l_train_test_set)):
    l_train_test_set[i]['Target_lag1'] = l_train_test_set[i]['Target'].shift(1)
    l_train_test_set[i]['Target_lag2'] = l_train_test_set[i]['Target'].shift(2)
    l_train_test_set[i]['Target_lag3'] = l_train_test_set[i]['Target'].shift(3)
    l_train_test_set[i]['Target_lag4'] = l_train_test_set[i]['Target'].shift(4)
    l_train_test_set[i]['Target_lag5'] = l_train_test_set[i]['Target'].shift(5)
    l_train_test_set[i]['Target_lag6'] = l_train_test_set[i]['Target'].shift(6)
    l_train_test_set[i]['Target_lag7'] = l_train_test_set[i]['Target'].shift(7)
    l_train_test_set[i]['Target_lag8'] = l_train_test_set[i]['Target'].shift(8)
    l_train_test_set[i]['Target_lag9'] = l_train_test_set[i]['Target'].shift(9)
    l_train_test_set[i]['Target_lag10'] = l_train_test_set[i]['Target'].shift(10)
    l_train_test_set[i]['Target_lag11'] = l_train_test_set[i]['Target'].shift(11)
    l_train_test_set[i]['Target_lag12'] = l_train_test_set[i]['Target'].shift(12)
    l_train_test_set[i]['Target_lag13'] = l_train_test_set[i]['Target'].shift(13)
    l_train_test_set[i]['Target_lag14'] = l_train_test_set[i]['Target'].shift(14)
    l_train_test_set[i]['Target_lag15'] = l_train_test_set[i]['Target'].shift(15)
    
    
    
    
```

## LightGBM Model (test)

```python
X = l_train_test_set[0].drop(columns = ['Target','datetime'])
y = l_train_test_set[0]['Target']
```

```python
# split the dataset into the training set and test set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 0)
```

```python
clf = lgb.LGBMRegressor()
clf.fit(X_train, y_train)
```

```python
y_pred= clf.predict(X_test)
```

```python
print(' R2: {0:0.4f}'. format(r2_score(y_pred, y_test)))
```

```python
pearsonr(y_pred, y_test)
```
